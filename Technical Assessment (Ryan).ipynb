```python
import pandas as pd

xls = pd.ExcelFile("Technical Test.xls")
sheet_names = xls.sheet_names
print(sheet_names)
```

```python
df = pd.read_excel("Technical Test.xls",sheet_name="Demographics")
```

```python
print(df.head(5))
df.info()
```

```python
#create New NRIC
print(df.NRIC.unique())
#remove space in NRIC
df["New NRIC"] = df["NRIC"].str.replace(" ", "")
df.head()
```

```python
def format_nric(nric):
    #check if the NRIC is in the SZ1234567 format
    if len(nric) == 9 and nric[1].isalpha() and nric[0].isalpha() and nric[-1].isdigit():
        nric = nric[0] + nric[2:] + nric[1]
    return nric

#apply the function to the "New NRIC" column
df["New NRIC"] = df["New NRIC"].apply(format_nric)
print(df["New NRIC"].unique())

#check the format of the New NRIC column
is_correct_format = df["New NRIC"].apply(lambda x: x[0].isalpha() and x[1:-1].isdigit() and x[-1].isalpha()).all()

print("All New NRIC are in S1234567Z format, True or False?") 
print("Answer:",is_correct_format)
```

```python
#gender coding
df["Coding - Gender"] = df["Gender"].replace({"M": 2, "F": 1})
df.head()
```

```python
#age group
#bins and labels for the age groups
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, float('inf')]
labels = ["G1", "G2", "G3", "G4", "G5", "G6", "G7", "G8", "G9", "G0"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels)

#study number
df["Study Number"] = df.groupby("Age Group").cumcount() + 1
df["Study Number"] = df["Age Group"].astype(str) + ' - ' + df["Study Number"].astype(str)

df.head()
```

```python
#extra information sheet
df2 = pd.read_excel("Technical Test.xls",sheet_name="Extra information")
print(df2)
print(df2.shape)
print(df2.columns)
```

```python
#study data
df3 = pd.read_excel("Technical Test.xls",sheet_name="Study Data")
print(df3)
print(df3.shape)
print(df3.columns)
```

```python
#columns to update in df3
columns_to_merge = ["Study Number", "New NRIC", "Gender", "Age", "Marital Status", "Address 1", "Address 2", "Contact Number", "Ethnic Group"]

#merge on "Old NRIC"
merged_df = df3.merge(df.add_suffix("_df"), left_on="Old NRIC", right_on="NRIC_df", how="left").drop_duplicates(subset="Old NRIC", keep="first")
print(merged_df.shape)

#update column values
for column in columns_to_merge:
    if column + '_df' in merged_df.columns:
        original_column = column
        merged_df[original_column] = merged_df[column + '_df']

#drop the suffixed columns
merged_df.drop(columns=[col for col in merged_df.columns if '_df' in col], inplace=True)
df3 = merged_df

#same process for df2
merged_df2 = df3.merge(df2.add_suffix("_df2"), left_on="Old NRIC", right_on="NRIC_df2", how="left")
for column in columns_to_merge:
    if column + "_df2" in merged_df2.columns:
        original_column = column
        merged_df2[original_column] = merged_df2[column + "_df2"]
merged_df2.drop(columns=[col for col in merged_df2.columns if "_df2" in col], inplace=True)
df3 = merged_df2

df3.head(10)
```

```python
#pivot tables
pivot_age_group_df = pd.pivot_table(df, index="Age Group", aggfunc="size").to_frame("Count")
pivot_gender_df = pd.pivot_table(df, index="Gender", aggfunc="size").to_frame("Count")
pivot_marital_status_df = pd.pivot_table(df, index="Marital Status", aggfunc="size").to_frame("Count")

#grand total
grand_total = pivot_age_group_df["Count"].sum() + pivot_gender_df["Count"].sum() + pivot_marital_status_df["Count"].sum()

#sub total
pivot_age_group_df.loc["Total", "Count"] = pivot_age_group_df["Count"].sum()
pivot_gender_df.loc["Total", "Count"] = pivot_gender_df["Count"].sum()
pivot_marital_status_df.loc["Total", "Count"] = pivot_marital_status_df["Count"].sum()

#combine
combined_pivot_df = pd.concat({
    "Age Group": pivot_age_group_df,
    "Gender": pivot_gender_df,
    "Marital Status": pivot_marital_status_df
})

#reset index and rename
combined_pivot_df.reset_index(inplace=True)
combined_pivot_df.columns = ["Variable", "Category", "Count"]

#add Grand Total at the end
combined_pivot_df = combined_pivot_df.append({"Variable": "Grand Total", "Category": "", "Count": grand_total}, ignore_index=True)

combined_pivot_df
```
